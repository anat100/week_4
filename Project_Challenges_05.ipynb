## Coin flip example

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Simulate coin flips and calculate estimated probabilities
num_trials = [10, 100, 1000, 10000]
estimated_probs = []

for n in num_trials:
    flips = np.random.choice(['H', 'T'], size=n)
    num_heads = np.sum(flips == 'H')
    estimated_probs.append(num_heads / n)

# Print the estimated probabilities
for n, prob in zip(num_trials, estimated_probs):
    print(f"Number of trials: {n}, Estimated Probability of Heads: {prob:.4f}")

# Plot the PMF of Binomial distribution
n = 10
p = 0.5
x = np.arange(0, n+1)
pmf_values = binom.pmf(x, n, p)

plt.bar(x, pmf_values, align='center', alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('PMF of Binomial Distribution (n=10, p=0.5)')
plt.show()
```

As you increase the number of trials (coin flips), you will likely observe that the estimated probability of getting a head gets closer and closer to the true probability of 0.5. This behavior is described by the Law of Large Numbers, which states that as the sample size increases, the sample mean (in this case, the estimated probability) approaches the population mean (true probability). In other words, with more trials, your estimate becomes more accurate and stable.
